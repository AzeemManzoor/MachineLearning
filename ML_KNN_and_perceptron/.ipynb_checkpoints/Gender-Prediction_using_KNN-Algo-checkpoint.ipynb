```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Machine Learning Data Collection - Sheet1.csv")
df
```

```python
x = df[['Height','Weight','Hairlength','Beard','Hijab','Physical Activities','Cooking','Own a pet','Hair Dye','Piercing','Contact Lens','Jewelery']]
y = df['Gender']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=13)
```

```python
knn.fit(X_train,y_train)
print(X_test)
```

```python
pred = knn.predict(X_test)
print(pred)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
print(KNeighborsClassifier.predict([170,130,4,1,0,1,1,0,0,0,1,0]),(df))
```
